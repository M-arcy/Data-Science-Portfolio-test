Here's a listing of the number of power boats registered in Florida (in the 1000's) for each year, 35 years worth of data. Tinker with the settings until you have created a histogram with 7 bars.

```r
library(ggplot2)
library(rcompanion)
library(tidyr)
```

```r
boats <- read.csv("L3P1.csv")
```

```r
head(boats)
```

```r
plotNormalHistogram(boats$Power.Boats..in.1000.s.)
```

```r
ggplot(boats, aes(x = Power.Boats..in.1000.s.)) + 
  geom_histogram(fill="#FF9999", color="black", 
                 binwidth = 100)
```

Part 5
Here's data tracking heart attacks treated at a hospital chain in a large US city. Use this data to create a line chart. Be sure to label the graph, and label both axes.

```r
hearts <- read.csv("L6Handson.csv")
```

```r
head(hearts)
```

```r
#Merging columns to have one date column.
HeartAttacks <- unite(hearts, Date, Month, Day, Year, sep=" ")
```

```r
#Checking Data Type
str(HeartAttacks)
```

```r
#Creating Line Chart
ggplot(HeartAttacks, aes(as.Date(Date, "%B %d, %Y"), Heart.Attacks))+
  geom_line() +
  ggtitle("Number of Heart Attacks from May to July in 2003") +
  xlab("Date") +
  ylab("Heart Attacks")
```
